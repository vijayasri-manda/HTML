```python
# Print the current column names to identify the correct spelling
print(df.columns)

# Correct the column name in the drop list. Assuming the correct name is 'Vehicle Size' based on common datasets.
df1=df.drop(["Engine Fuel Type",
            "Market Category",
            'Vehicle Style',
            'Popularity',
            'Number of Doors',
            'Vehicle Size'],axis =1) # Corrected 'Vehicles Size' to 'Vehicle Size'
```

```python
df.head()
```

```python
df1=df1.rename(columns={'Make':'Brand',
                        'Engine HP':'HP',
                        'Engine Cylinders':'Cylinders',
                        'Transmission Type':'Transmission'})
```

```python
df1=df1.drop_duplicates()
```
